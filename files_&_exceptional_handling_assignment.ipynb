 3. What is the purpose of the finally block in exception handling?
    - The purpose of the finally block in exception handling is to ensure that a specific block of code, typically for resource cleanup, always executes regardless of whether an exception is thrown or not, or even if an exception is thrown and not caught.

```python
def divide(num1, num2):
    return num1 / num2

try:
    # Code that may cause an exception
    result = divide(10, 0)  # Attempting to divide by zero
    print("Result:", result)  # This line won't be executed
except ZeroDivisionError as e:
    # Handling the exception
    print("An error occurred:", e)
finally:
    # finally block
    print("\nFinally block executed, performing cleanup tasks.")
```

     - Python uses two strategies for memory allocation:
        -> Reference counting
        -> Garbage collection
     - Reference counting:
     Python and various other programming languages employ reference counting, a memory management approach, to automatically manage memory by tracking how many times an object is referenced. A reference count, or the number of references that point to an object, is a property of each object in the Python language. When an object’s reference count reaches zero, it becomes un-referenceable and its memory can be freed up.


```python
# Create an object
x = [1, 2, 3]

# Increment reference count
y = x

# Decrement reference count
y = None

# Create two objects that refer to each other
x = [1, 2, 3]
y = [4, 5, 6]
x.append(y)
y.append(x)


import sys

# Create an object
x = [1, 2, 3]

# Get reference count
ref_count = sys.getrefcount(x)

print("Reference count of x:", ref_count)
```

 23. What are the challenges associated with memory management in Python?
     - Python's automatic memory management, while simplifying development, presents challenges like potential memory leaks due to circular references and the need for memory profiling tools to identify inefficiencies.
     - Here's a more detailed breakdown of the challenges:
1. Circular References and Memory Leaks:
Problem:
Python uses reference counting for memory management, but it can struggle with circular references (where two or more objects point to each other).
Explanation:
If two objects are circularly referencing each other, their reference counts never drop to zero, preventing the garbage collector from reclaiming the memory they occupy, leading to memory leaks.
Solution:
Be mindful of circular references and use weak references (using the weakref module) when appropriate to avoid this issue.
2. Memory Profiling and Optimization:
Problem:
Python's automatic memory management can make it difficult to pinpoint memory usage issues, such as memory leaks or inefficient memory allocation.
Solution:
Use memory profiling tools (like memory_profiler or objgraph) to analyze memory usage and identify potential problems.
3. Memory Consumption with Large Datasets:
Problem:
Python's dynamic memory allocation can lead to high memory consumption when working with large datasets, potentially causing performance issues or crashes.
Solution:
Consider using memory-efficient data structures, chunking data processing, or using libraries like Dask for distributed computing to handle large datasets.
4. Deallocation and Fragmentation:
Problem:
Deallocating memory can be expensive and potentially lead to fragmentation, where memory is split into unusable chunks.
Explanation:
Just like allocation, deallocation requires telling the operating system and MMU that the process is done with the memory, which can be slow.
Solution:
Be mindful of memory allocation and deallocation patterns to minimize fragmentation and memory leaks.
5. Concurrency and the Global Interpreter Lock (GIL):
Problem:
Python's GIL can limit true parallelism in CPU-bound tasks, making it challenging to fully utilize multi-core processors.
Explanation:
The GIL allows only one thread to hold control of the Python interpreter at any given time, which can lead to bottlenecks in multithreaded applications.
Solution:
Consider using asynchronous programming with asyncio for I/O-bound tasks, or use libraries like multiprocessing for CPU-bound tasks that can bypass the GIL.
6. Memory Safety:
Problem:
Python's automatic memory management can lead to memory leaks and performance issues if not managed correctly.
Explanation:
Python uses garbage collection to automatically free up memory, which can lead to memory leaks and performance issues if not managed correctly.
Solution:
Be mindful of memory usage and use memory profiling tools to identify potential problems.

 21. Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
# Open the file in read mode
text = open("sample.txt", "r")

# Create an empty dictionary
d = dict()

# Loop through each line of the file
for line in text:
	# Remove the leading spaces and newline character
	line = line.strip()

	# Convert the characters in line to
	# lowercase to avoid case mismatch
	line = line.lower()

	# Split the line into words
	words = line.split(" ")


	# Iterate over each word in line
	for word in words:
		# Check if the word is already in dictionary
		if word in d:
			# Increment count of word by 1
			d[word] = d[word] + 1
		else:
			# Add the word to dictionary with count 1
			d[word] = 1

# Print the contents of dictionary
for key in list(d.keys()):
	print(key, ":", d[key])
```
